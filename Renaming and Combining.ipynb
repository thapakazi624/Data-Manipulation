# Imports

```python
import pandas as pd
wine_data = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)
pd.set_option('display.max_rows', 5)
```

```python
# Renaming a column
wine_data.rename(columns={'points':'score'})
```

```python
# renaming index
wine_data.rename(index={0: 'first', 1:'second'})
```

```python
# renaming axis
wine_data.rename_axis('wines', axis='rows').rename_axis('fields', axis='columns')
```

# Combining

```python
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
pd.concat([df, df2])
```

```python
# using join
df.join(df2, lsuffix='_df', rsuffix='_df2')
```
