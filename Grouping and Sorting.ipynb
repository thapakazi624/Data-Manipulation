# Imports

```python
import pandas as pd
wine_data = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)
pd.set_option('display.max_rows', 5)
```

```python
wine_data.head()
```

```python
# use of groupby and count methods
wine_data.groupby('points').points.count()
```

```python
# use of groupby and min methods
wine_data.groupby('points').points.min()
```

```python
# groupby and apply with lambda function
wine_data.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# groupby more than one columns and apply with lambda function
wine_data.groupby(['country','province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# use of agg() with groupby
wine_data.groupby(['country']).price.agg([len, min, max])
```

# Multi Indexes

```python
countries_data = wine_data.groupby(['country','province']).description.agg([len])
```

```python
countries_data
```

```python
# we can confirm multiple index by checking the type
mi = countries_data.index
type(mi)
```

```python
# converting to old form
countries_data.reset_index()
```

# Sorting

```python
# sorting using sort_values 
countries_data = countries_data.reset_index()
countries_data.sort_values(by='len')
```

```python
# descending order
countries_data.sort_values(by='len', ascending=False)
```

```python
# sorting on the basis of index
countries_data.sort_index()
```

```python
# sorting more than one column
countries_data.sort_values(by=['province','len'])
```
