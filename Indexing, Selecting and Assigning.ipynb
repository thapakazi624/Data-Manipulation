```python
data_flight.iloc[:, 0]
```

```python
data_flight.iloc[:3, 0]
```

```python
data_flight.iloc[1:3, 0]
```

```python
data_flight.iloc[[0,1,2], 0]
```

```python
# counting from end of the dataframe
data_flight.iloc[-5:]
```

```python
# label based selection
data_flight.loc[:,'ARG']
```

```python
# selecting multiple columns
data_flight.loc[:, ['ARG','BRA']]
```

```python
# Manipulating the Index
data_flight.set_index('ARG')
```

```python
# conditional selection
data_flight.BRA == 1.0
```

```python
data_flight.loc[data_flight.BRA == 1.0]
```

```python
# and operation
data_flight.loc[(data_flight.BRA == 1.0) & (data_flight.FRA == 6.0)]
```

```python
# or operation
data_flight.loc[(data_flight.BRA == 1.0) | (data_flight.FRA == 6.0)]
```

```python
# isin operation
data_flight.loc[data_flight.FRA.isin([1.0,6.0])]
```

```python
# is null and not null
data_flight.loc[data_flight.ARG.notnull()]
```
